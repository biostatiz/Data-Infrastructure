```python
idlist=[]
createdatlist=[]
textlist=[]
for tweets in tweepy.Cursor(api.user_timeline, id='realDonaldTrump').items(200):
    idlist.append(tweets.id)
    createdatlist.append(tweets.created_at)
    textlist.append(tweets.text)
tweetdata = {
    "id":idlist,
    "createdat":createdatlist,
    "text":textlist
}
```

```python
tweetdf=pd.DataFrame(tweetdata)
```
